```python
#imporing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
# importing the dataset for train
df_train = pd.read_csv(r"C:\\Users\\HP.000\\Desktop\\Program\\Black Friday\\train.csv")
df_train.head()
```

```python
# importing the dataset for test
df_test = pd.read_csv(r"C:\\Users\\HP.000\\Desktop\\Program\\Black Friday\\test.csv")
df_test.head()
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
## Merge both train and test dataset
df = pd.concat([df_train, df_test], ignore_index=True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Deleting unused column
df.drop(["User_ID"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Handling categorical feature Gender.

df["Gender"] = df["Gender"].map({"F":0,"M":1}) #map methods use to convert variable into given condition
df.head()
```

```python
# Handling categorical feature Age

df["Age"].sort_values().unique()
```

```python
# using LabelEncoder converting categorical feature "Age" into numeric.

from sklearn import preprocessing
 
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
 
# Encode labels in column 'species'.
df['Age']= label_encoder.fit_transform(df['Age'])
 
df['Age'].sort_values().unique()
```

```python
# df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
# fixing categorical City_Category

df_city = pd.get_dummies(df["City_Category"], drop_first = True)
```

```python
df_city
```

```python
# Concatening new columns to existing dataframe

df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
#Converting boolean value to numeric value i.e True = 1 & False = 0

df['B'] = df['B'].astype(int)
df['C'] = df['C'].astype(int)
```

```python
df.head()
```

```python
# droping City_Category feature as there is more used

df.drop(columns = "City_Category", inplace = True)
df.head()
```
